```python
# Exercise: Parsing with BS4

# Retrieve the content from http://quotes.toscrape.com/
# Retrieve the first quote's text.
# Retrieve the second quote's author.
# Retrieve all of the quotes, iterate over them, and print them.
```

```python
import requests

url = 'https://quotes.toscrape.com'

r = requests.get(url)
```

```python
print(r.content.decode())
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
print(soup.get_text())
```

```python
soup.select('div.quote')[0]
```

```python
soup.find('div', attrs={'class':'quote'})
```

```python
# first quote's text
soup.select('div.quote span.text')[0].text
```

```python
# second quote's author
soup.select('div.quote small.author')[1].text
```

```python
for one_quote_div in soup.select('div.quote'):
    for one_child in one_quote_div.children:
        print(one_child)
```

```python
for one_quote_div in soup.select('div.quote'):
    for one_child in one_quote_div.children:
        if one_child.name == 'span':
            print(one_child.text.replace('(about)', '').strip())
```
